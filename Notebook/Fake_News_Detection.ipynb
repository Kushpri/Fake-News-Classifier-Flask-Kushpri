```python
import re
import string
import pandas as pd
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r"\W", " ", text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text
```

```python
def output_label(n):
    if int(n) == 0:
        return "Fake News"
    elif int(n) == 1:
        return "Not A Fake News"
    return "Unknown"

def manual_testing(news_text, return_probs=False, majority_vote=True):
    """
    Take a single news string and run it through the trained models.
    - news_text: str (one news article)
    - return_probs: if True, also return predict_proba for each model (if available)
    - majority_vote: if True, include a final label by majority vote of models
    Returns: dict with model predictions, optional probabilities, and final label (if majority_vote=True)
    """
    # Sanity checks
    required = ["vectorization", "LR", "DT", "GBC", "RFC", "wordopt"]
    for r in required:
        if r not in globals():
            raise NameError(f"Required object '{r}' not found. Define/train it before calling manual_testing().")

    # Prepare input
    if not isinstance(news_text, str):
        raise ValueError("news_text must be a string.")
    new_df = pd.DataFrame({"text": [news_text]})
    new_df["text"] = new_df["text"].apply(wordopt)

    # Vectorize
    new_xv = vectorization.transform(new_df["text"])

    # Predict (take first element since input is single)
    pred_lr = LR.predict(new_xv)[0]
    pred_dt = DT.predict(new_xv)[0]
    pred_gbc = GBC.predict(new_xv)[0]
    pred_rfc = RFC.predict(new_xv)[0]

    results = {
        "LR": output_label(pred_lr),
        "DT": output_label(pred_dt),
        "GBC": output_label(pred_gbc),
        "RFC": output_label(pred_rfc),
    }

    # optionally get probabilities (if model supports predict_proba)
    if return_probs:
        probs = {}
        for name, model in [("LR", LR), ("DT", DT), ("GBC", GBC), ("RFC", RFC)]:
            if hasattr(model, "predict_proba"):
                p = model.predict_proba(new_xv)[0]  # [prob_of_class0, prob_of_class1]
                probs[name] = {"fake_prob": float(p[0]), "not_fake_prob": float(p[1])}
            else:
                probs[name] = None
        results["probs"] = probs

    # majority vote final label (simple)
    if majority_vote:
        votes = [int(pred_lr), int(pred_dt), int(pred_gbc), int(pred_rfc)]
        # majority: 0 => Fake, 1 => Not Fake
        final = 1 if sum(votes) > (len(votes) / 2) else 0
        results["final_label"] = output_label(final)
        results["votes"] = {"fake": votes.count(0), "not_fake": votes.count(1)}

    # print nicely
    print(
        f"\nLR Prediction: {results['LR']}\n"
        f"DT Prediction: {results['DT']}\n"
        f"GBC Prediction: {results['GBC']}\n"
        f"RFC Prediction: {results['RFC']}\n"
    )
    if majority_vote:
        print(f"Majority vote -> Final Label: {results['final_label']} (votes: {results['votes']})\n")
    if return_probs:
        print("Probabilities (if available):", results.get("probs"))

    return results
```
